```python
df_info_norm = df_info[df_info['classification'] == 'norm'].copy()
institution_counts_norm = df_info_norm['institution_id'].value_counts()
hospitals_to_keep = institution_counts_norm[institution_counts_norm >= 100].index

df_info_filtered = df_info[df_info['institution_id'].isin(hospitals_to_keep)]
df_info_filtered.to_csv(ELM19_info_filtered_path, index=False)
```
